```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

### DATA CLEANING AND PROCESSING

```python
data = pd.read_csv("superstore_dataset.csv", encoding="latin1")
pd.set_option('display.max_columns', None)
data.head(5)
```

```python
data.isna().sum()
data.nunique()
```

```python
## Data Visualization
```

```python
**Which region generates the most revenue?**
```

```python
Revenue_Per_Region = data.groupby("Region")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,8))
plt.pie(Revenue_Per_Region.values, labels=Revenue_Per_Region.index, autopct='%1.1f%%', startangle=140)
plt.title("Revenue by Region")
plt.show()
```

```python
Revenue_Per_Region = data.groupby("Region")["Sales"].sum().sort_values(ascending = False)
print(Revenue_Per_Region)
Revenue_Per_Region = Revenue_Per_Region.reset_index()
```

```python
plt.figure(figsize  = (16,6))

sns.barplot(x='Region', y='Sales', data = Revenue_Per_Region, width = 0.5, palette = ["green", "red"])
plt.title("Sales BY Region")
plt.show()
```

```python
**which category is more profitable?**
```

```python
profit_per_category = data.groupby("Category")["Profit"].sum()
print(profit_per_category)
```

```python
plt.figure(figsize = (5,5))
plt.bar( profit_per_category.index, profit_per_category.values, color = "green")
plt.xlabel("Categories")
plt.ylabel("Profit")
plt.title("Profit Per Category Product")
```
